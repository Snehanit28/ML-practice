## EDA(Explorartory Data Analysis) Using Univariate Analysis ##

```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
df.head()
```

# 1. Catagorical Data #

**a. Countplot**

```python
sns.countplot(x=df['Pclass'])  #countplot of passenger class
# .countplot()->  counts the value of survived and plot as bar char
```

```python
df['Survived'].value_counts().plot(kind='bar')     #same work as sns.countplot
# .value_counts()->  counts the value of survived ,
# .plot(kind='bar')->  plot the counts as a bar chart
```

**b. PieChart** 

```python
df['Embarked'].value_counts().plot(kind='pie')   #pie chart of embarked
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct = '%.2f')
```

# 2. Numerical Data #

**a. Histogram**

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=20)  #histogram of age ,if we want to set number of bins we can use bins parameter and set its value ,if we dont set it default value is 10
# plt.xlabel('Age')    # label for x axis
# plt.ylabel('Count')  # label for y axis
# plt.title('Age Distribution')  # title of the histogram
```

**b. Distplot**

```python
# sns.displot(df['Age'],bins=20,kde=True)  #displot of age with kde line in this x axis we have age and in y axis we have count
sns.histplot(df['Age'],bins=20,kde=True,stat='density')  #histplot of age with kde line and density on y axis
```

**c. Boxplot**

```python
sns.boxplot(x=df['Age'])   
```
